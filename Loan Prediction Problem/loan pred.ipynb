```python
df= df[["Dependents", "Education", 'Self_Employed', 'ApplicantIncome', 'LoanAmount', 'Loan_Status' ]]
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
new_values = pd.get_dummies(df[['Education', 'Self_Employed', 'Loan_Status']], drop_first=True)
```

```python
df= pd.concat([df[['Dependents', 'ApplicantIncome', 'LoanAmount']],new_values],axis=1)
```

```python
df.head()
```

```python
df['Dependents'] = df['Dependents'].replace(to_replace='3+',value='4').astype(int)
```

```python
df.info()
```

```python
x= df.drop(['Loan_Status_Y'] ,axis=1)
y=df['Loan_Status_Y']
```

```python
x = np.array(x)
y = np.array(y)
```

## logisticregression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model= lr.fit(x,y)
model.score(x,y)
```

## randomforest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
model = clf.fit(x,y)
model.score(x,y)
```

## naive bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
model = nb.fit(x,y)
model.score(x,y)
```

## svm

```python
from sklearn import svm
clf = svm.SVC()
model = clf.fit(x,y)
model.score(x,y)
```

## decision tree

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
model = clf.fit(x,y)
model.score(x,y)
```

## random forest classification 

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
model = clf.fit(x,y)
model.score(x,y)
```
